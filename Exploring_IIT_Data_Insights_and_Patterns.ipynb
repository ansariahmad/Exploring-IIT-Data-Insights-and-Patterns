### Renaming each column with its abbreviation.

In this section name of all columns will be renamed with their abbreviatiosn/short form. Any time you are failed to understand the the column name, visit this table.

| Previous Name | New Name |
|----------|----------|
|   Student Id   |   id   |
|   Student Name   |   name   |
|   Date of Birth   |   dob   |
|   Field of Study   |   field   |
|   Year of Admission   |   admission_year   |

```python
# want to read more about pd.DataFrame.rename? Uncomment the next line.
# print(pd.DataFrame.rename.__doc__)
```

```python
# a dictionary mapper
column_mapper = {
    "Student Id" : "id",
    "Student Name" : "name",
    "Date of Birth" : "dob",
    "Field of Study" : "field",
    "Year of Admission" : "admission_year",
    "Expected Year of Graduation" : "exp_grad_year",
    "Current Semester" : "semester",
    "Specialization" : "specialization",
    "Fees" : "fees",
    "Discount on Fees" : "fee_discount"
}
# passing dictionary mapper to rename method.
iit.rename(column_mapper, axis=1, inplace=True)
```

### Change the datatype of `dob` column to datetime

```python
# want to read more about pd.datetime? Uncomment the following line.
# print(pd.to_datetime.__doc__)
```

```python
iit["dob"] = pd.to_datetime(iit["dob"])
```

### Feature creation

Creating `age` column.

```python
# want to read more about pd.Timestamp? Uncomment the following line.
# print(pd.Timestamp.__doc__)
```

```python
# `age` col has been created by subtracting the current date and the date of birth of student.
iit["age"] = (pd.Timestamp(datetime.datetime(2023, 6, 12)) - iit["dob"]).astype('<m8[Y]').astype('int')
```

### Q1: What is the distribution of fees?

```python
# want to read more about sns.kdeplot? Uncomment the following line.
# print(sns.kdeplot.__doc__)
```

```python
sns.kdeplot(data=iit, x="fees")     # creating kernel density plot
plt.xlabel("Fees")                  # setting x-axis label
plt.ylabel("Probability Density")   # setting y-axis label
plt.title("Distribution of fees")   # setting title of plot
plt.plot()
```

It shows that Fees col is uniformly distributed.

Uniform distribution refers to a type of probability distribution in which all outcomes are equally likely.

### Q2: Is there any correlation between fees and discount on fees?

```python
# want to read more about sns.scatterplot? Uncomment the following line.
# print(sns.scatterplot.__doc__)
```

```python
# want to read more about plt.text? Uncomment the following line.
# print(plt.text.__doc__)
```

```python
corr = iit[["fees", "fee_discount"]].corr()["fee_discount"][0]      # finding correlation between fees and discount on fee
sns.scatterplot(data=iit, x="fees", y="fee_discount")               # creating a scatterplot
plt.xlabel("Fees")
plt.ylabel("Discount on Fees")
plt.title("Scatter plot displaying correlation between Fees and Discount on Fees")
plt.text(60000, 35000, f"Correlation: {corr:.2}")                   # displaying text on the plot
plt.show()
```

### Conclusion:
The fees and fee_discount exhibit a moderate correlation with each other, as indicated by a correlation value of 0.49.

#### Q3: Any changes in fees over time.

```python
fee_ = iit.groupby("admission_year")["fees"].agg(["mean", "median"])
fee_
```

Let's visualize the output.

```python
sns.lineplot(data=fee_, x="admission_year", y="mean", label="mean", marker="o")       # creating lineplot for mean
sns.lineplot(data=fee_, x="admission_year", y="median", label="median", marker="o")   # creating lineplot for median
plt.xlabel("Admission Year")
plt.ylabel("Fees")
plt.legend(bbox_to_anchor=(1.225, 1.02))
```

### Conclusion:
There is a significant decrease in fees at IIT in the year 2018, followed by a period of improvement. However, there is another decline in fees at IIT in the year 2022.

#### Q4: How the fees of different specializations within each field of study changes with time?

```python
# increment in fees of specialization with year
fee_year = iit.groupby(["admission_year", "field", "specialization"])["fees"].mean().reset_index()
fee_year.sample(4)
```

The above dataframe displays the adission year, field, specialization and the fees of that time. But it is quite difficult to make a conclusion. Let's draw a lineplot.

```python
g = sns.relplot(data=iit, x="admission_year", y="fees", col="field", row="specialization", kind="line")     # creating a lineplot
# setting x-axis & y-axis of every subplot
for ax in g.axes.flat:
    ax.set_xlabel('Admission Year')
    ax.set_ylabel('Fees')
plt.plot()
```

### Conclusion:

It is evident that the fees for many specializations within each field of study have decreased over time. In some cases, there have been drastic changes in fees, such as in the field of computer science with the specialization in web development. However, we also observe that fees have remained consistent over time for certain specializations, such as Data Science in Civil Engineering. Additionally, there are notable improvements, like the specialization of data science in the field of chemical engineering.

#### Q5: Let's analyze how the number of students has changed over time for each specialization within every field of study.


```python
temp = iit.groupby(["admission_year", "field", "specialization"]).size().reset_index()
temp.rename({0:"student_count"}, axis=1, inplace=True)
temp.head()
```

The above dataframe displays the adission year, field, specialization and the number of students enrolled that time. But it is quite difficult to make a conclusion. Let's draw a lineplot.

```python
g = sns.relplot(data=temp, x="admission_year", y="student_count", col="field", row="specialization", kind="line")   # creating a lineplot
# setting x-axis & y-axis of every subplot
for ax in g.axes.flat:
    ax.set_xlabel('Admission Year')
    ax.set_ylabel('No. of Students')

plt.show()
```

### Conclusion:

Upon analyzing the plot, we observe several noteworthy trends. Firstly, the specialization of AI in the field of electrical engineering experiences a decline in the number of students over time. Conversely, there is a substantial increase in the number of students pursuing network security in the field of Civil Engineering.

## Inferences and Conclusion

During the exploratory data analysis (EDA), several findings were observed:

- Variation in Fee Discounts: Some students receive higher fee discounts compared to others, potentially due to their high CGPA or admission test scores.

- Uniform Distribution of Fees: The fees column follows a uniform distribution, indicating a relatively even distribution of fee values.

- Correlation between Fees and Discounts: There is a moderate positive correlation of 0.50 between fees and the discount on fees, suggesting that as fees increase, the discount also tends to increase.

- Fee Trends over Time: Fees experienced a decrease in 2018, followed by an increase until 2020, and then a continuous decrease from 2021 to 2022.

- Average Fees by Specialization: On average, fees for many specializations within each field decreased over time, while some specializations showed an increase in fees.

- Enrollment Trends by Specialization: In the field of Electrical Engineering, the specialization of Artificial Intelligence exhibited a reduction in the number of students over time. Conversely, in Computer Science, the specialization of Web Development showed an increment in the number of students over time.

These insights provide valuable information about fee patterns, trends, and student enrollment changes across different specializations within each field.
